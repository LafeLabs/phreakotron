```python
import numpy as np
from IPython.display import Audio

# Set the sample rate to 44.1 kHz
sample_rate = 44100

# Set the duration to 1 second
duration = 1.0

# Set the frequency to 2600 Hz
frequency = 1400.0

# Generate the sine wave
t = np.linspace(0, duration, int(sample_rate*duration), False)
waveform = np.sin(2*np.pi*frequency*t)

# Create the Audio object and play the tone
Audio(waveform, rate=sample_rate, autoplay=True)
```
